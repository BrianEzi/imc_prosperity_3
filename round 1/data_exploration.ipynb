```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filepath = 'round-1-island-data-bottle/round-1-island-data-bottle/'
price_day_filename = lambda x :  f'prices_round_1_day_{x}.csv'
trade_day_filename = lambda x :  f'trades_round_1_day_{x}.csv'

prices_df = pd.read_csv(filepath + price_day_filename(-2), delimiter=';')
trades_df = pd.read_csv(filepath + trade_day_filename(-2), delimiter=';')

# Filter for squid ink and kelp/resin
squid_prices = prices_df[prices_df["product"] == "SQUID_INK"]
kelp_prices = prices_df[prices_df["product"] == "KELP"]
resin_prices = prices_df[prices_df["product"] == "RAINFOREST_RESIN"]

# Calculate midprice for squid ink (already exists as mid_price column)
squid_mid = squid_prices.set_index("timestamp")["mid_price"]

# Calculate net trade flow for kelp: sum(buy_volume - abs(sell_volume)) for each timestamp
kelp_trades = trades_df[trades_df["symbol"] == "KELP"]
kelp_trades["net_flow"] = kelp_trades.apply(lambda row: row["quantity"] if row["buyer"] == "SUBMISSION" else -row["quantity"], axis=1)
kelp_net_flow = kelp_trades.groupby("timestamp")["net_flow"].sum()

# Align and resample the series (depending on your timestamp granularity)
# For demonstration, assume timestamps match and convert to series
squid_series = squid_mid.astype(float)
kelp_series = kelp_net_flow.astype(float)

# Compute cross-correlation
max_lag = 10  # Check lags between -10 and 10 timesteps
lags = range(-max_lag, max_lag+1)
corrs = [squid_series.corr(kelp_series.shift(lag)) for lag in lags]

plt.plot(lags, corrs)
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.title("Cross-Correlation: Squid Ink midprice vs Kelp net trade flow")
plt.show()

# The lag with highest absolute correlation might signal the reaction time of the unknown actor.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.stattools as smt
from statsmodels.tsa.stattools import grangercausalitytests

# -------------------------------
# STEP 1: Load the Data
# -------------------------------

# Adjust the file paths as needed.
day = -0

prices_df = pd.read_csv(filepath + price_day_filename(day), delimiter=';')
trades_df = pd.read_csv(filepath + trade_day_filename(day), delimiter=';')
# -------------------------------
# STEP 2: Filter Data for Each Asset
# -------------------------------

# Squid Ink prices: assuming the column names exactly as provided
squid_prices = prices_df[prices_df["product"] == "SQUID_INK"]

# Kelp and Rainforest Resin prices
kelp_prices = prices_df[prices_df["product"] == "KELP"]
resin_prices = prices_df[prices_df["product"] == "RAINFOREST_RESIN"]

# -------------------------------
# STEP 3: Build Time Series
# -------------------------------

# For Squid Ink we use the provided mid_price as a time series.
# Assume timestamp is our index.
squid_series = squid_prices.set_index("timestamp")["mid_price"].astype(float)

# For Kelp and Resin, we want to create a net trade flow time series.
# For trades, we assume that a trade is positive when it is a buy for your submission,
# but here, since we don't have explicit labels, we'll assume a trade with positive quantity is a buy.
# Adjust the logic if buyer/seller fields are available.
def create_net_trade_flow(symbol):
    df_symbol = trades_df[trades_df["symbol"] == symbol].copy()
    # Convert quantity to integer
    df_symbol["quantity"] = df_symbol["quantity"].astype(int)
    # In our simple example, we'll assume positive quantity = buy, negative = sell.
    # If buyer and seller indicators are available, you can adjust this logic.
    # For now, we sum quantities per timestamp.
    net_flow = df_symbol.groupby("timestamp")["quantity"].sum()
    return net_flow

kelp_net_flow = create_net_trade_flow("KELP")
resin_net_flow = create_net_trade_flow("RAINFOREST_RESIN")

# For consistency, convert index to numeric and sort.
squid_series = squid_series.sort_index()
kelp_net_flow = kelp_net_flow.sort_index()
resin_net_flow = resin_net_flow.sort_index()

# Optionally, resample if your timestamps are not continuous (here we assume they are comparable).
# For example, if timestamp represents seconds or ticks you could fill missing values via:
# squid_series = squid_series.asfreq('S').fillna(method="ffill")
# But for now we assume your timestamps line up as desired.

# -------------------------------
# STEP 4: Compute Lagged Cross-Correlation
# -------------------------------

def compute_cross_correlation(series1, series2, max_lag=10):
    """
    Compute cross-correlation for lags between -max_lag and +max_lag.
    Returns lags and correlation coefficients.
    """
    lags = np.arange(-max_lag, max_lag + 1)
    corrs = []
    for lag in lags:
        if lag < 0:
            shifted = series2.shift(-lag)
            corr = series1.corr(shifted)
        else:
            shifted = series2.shift(lag)
            corr = series1.corr(shifted)
        corrs.append(corr)
    return lags, corrs

max_lag = 10

# Compute cross-correlation for Kelp vs Squid Ink
lags_kelp, corrs_kelp = compute_cross_correlation(squid_series, kelp_net_flow, max_lag)
# Compute cross-correlation for Resin vs Squid Ink
lags_resin, corrs_resin = compute_cross_correlation(squid_series, resin_net_flow, max_lag)

# Plot the results
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(lags_kelp, corrs_kelp, marker="o")
plt.title("Cross-Correlation: Squid Ink vs. Kelp Net Flow")
plt.xlabel("Lag (timestamps)")
plt.ylabel("Correlation Coefficient")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(lags_resin, corrs_resin, marker="o", color="orange")
plt.title("Cross-Correlation: Squid Ink vs. Resin Net Flow")
plt.xlabel("Lag (timestamps)")
plt.ylabel("Correlation Coefficient")
plt.grid(True)

plt.tight_layout()
plt.show()

# -------------------------------
# STEP 5: Granger Causality Test
# -------------------------------
# We can run a Granger causality test to see if past values of kelp_net_flow predict squid_series.
# For Granger causality, we need to align the two time series into one DataFrame.

# For demonstration, we test for Kelp net flow
df_test = pd.concat([squid_series, kelp_net_flow], axis=1)
df_test.columns = ["squid_mid", "kelp_net_flow"]
df_test = df_test.dropna()

# We test for lags up to max_lag (adjust as necessary)
print("Granger Causality Test Results (Kelp -> Squid Ink):")
granger_result = grangercausalitytests(df_test[["squid_mid", "kelp_net_flow"]], max_lag, verbose=True)

# You can repeat a similar test for Resin if desired.
```
